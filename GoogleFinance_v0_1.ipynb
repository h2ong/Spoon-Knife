### Accessing Google Sheets in Python

Run the cell below to authenticate your account and connect to your Google Sheets. This will prompt you to log in and grant permissions.

```python
# Set Up Authentication in Google Colab

from google.colab import auth
import gspread
from google.auth import default

# Authenticate user and get credentials - from Colab account, which should be my
# Google account
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
# Create a New Google Sheet

# Create a new spreadsheet with a specific name in exisiting folder
spreadsheet_name = 'My New Colab Sheet'
folder_id = '1xAyI3iMucpycgZuAxG708oESdz6GStMN' # 'Trading' folder
sh = gc.create(spreadsheet_name, folder_id)
```

```python
# Open Exisiting Google Sheet
# Use gc.open
```

```python
# Step 4: Run Functions/Formulas on the Sheet

worksheet = sh.sheet1

worksheet.update([['Hello world']], 'A2')
worksheet.update([[24]], 'A1')
worksheet.update([['=SUM(B1:B4)']], 'A3', raw=False)
worksheet.update([['=SUM(B1:B4)']], 'A4', raw=True) # shows raw formula as text

# worksheet2 = sh.sheet2 # Doesn't work
worksheet2 = sh.add_worksheet(title="Sheet2", rows="100", cols="20")
worksheet2.update([['Hello world']], 'A1')
worksheet2.update([['=GOOGLEFINANCE("NYSEARCA:VTI", "all", DATE(2001,1,1), TODAY(), "DAILY")']], 'A1', raw=False)

worksheet3 = sh.add_worksheet(title="Sheet3", rows="100", cols="20")
# Needed escape character to add single quotation
worksheet3.update([['=QUERY(GOOGLEFINANCE("NYSEARCA:VTI", "all", DATE(2001,1,1), TODAY(), "DAILY"), "SELECT * format Col1 \'yyyy-MM-dd\'", 1)']], 'A1', raw=False)
```
